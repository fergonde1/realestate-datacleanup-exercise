#### Ejercicio 08. Trazar el histograma de los precios para la población (level5 column) de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

ds_Arroyomolinos = ds[ds['level5'] == 'Arroyomolinos (Madrid)']

media=ds_Arroyomolinos['price'].mean()
moda=ds_Arroyomolinos['price'].mode().iloc[0]
mediana=ds_Arroyomolinos['price'].median()

print(f"La media de Arroyomolinos es {int(media)}")
print(f"La moda de Arroyomolinos es {int(moda)}")
print(f"La mediana de Arroyomolinos es {int(mediana)}")

plt.figure(figsize = (10, 5))
plt.hist(ds_Arroyomolinos['price'], bins = 30, alpha = 0.7)
plt.title("Histogram")
plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label=f'Media: {int(media)}')
plt.axvline(moda, color='green', linestyle='dashed', linewidth=1, label=f'Moda: {int(moda)}')
plt.axvline(mediana, color='blue', linestyle='dashed', linewidth=1, label=f'Mediana: {int(mediana)}')
plt.legend()
plt.show()
```

El histograma presenta una distribución normal, similar a una campana de Gauss, con una moda estadística cercana a la media aunque está ligeramente desplazada hacia el eje izquierdo, al ser la media superior a la mediana y moda.

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

filtro = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

subconjunto = ds[ds['level5'].isin(filtro)]
mean_price = subconjunto['price'].mean()
std_price = subconjunto['price'].std()
subconjunto_normalized = subconjunto.copy()
subconjunto_normalized['price'] = (subconjunto['price'] - mean_price) / std_price
plt.figure(figsize=(10, 6))
for valor in filtro:
    datos_valor = subconjunto_normalized[subconjunto_normalized['level5'] == valor]['price']
    plt.hist(datos_valor, bins=20, alpha=0.5, label=valor)
plt.title('Histograma Normalizado de Precios para Valores de level5')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Salvo Getafe todas las distribuciones están desplazadas hacia el eje izquierdo, lo que significa que presentan un gran conjunto de casas por debajo del precio medio. Getafe sí está más distribuido.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in the question 14 (★★☆)

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

filtro = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

subconjunto = ds[ds['level5'].isin(filtro)].copy()
subconjunto.loc[:, 'pps'] = subconjunto['price'] / subconjunto['surface']
print(subconjunto[subconjunto['level5'] == 'Alcorcón']['pps'].mean())
print(subconjunto[subconjunto['level5'] == 'Getafe']['pps'].mean())
```

```python
import pandas as pd
from ipyleaflet import Map, basemaps, Marker

ds = pd.read_csv('assets/real_estate.csv', sep=';')

filtro = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
subconjunto = ds[ds['level5'].isin(filtro)]
coordenadas = subconjunto[['latitude', 'longitude']].to_dict('records')
mapa = Map(center=(40.3, -3.8), zoom=10, min_zoom=1, max_zoom=20, basemap=basemaps.Stamen.Terrain)
for coordenada in coordenadas:
    marcador = Marker(location=(coordenada['latitude'], coordenada['longitude']))
    mapa.add_layer(marcador)
mapa
```
